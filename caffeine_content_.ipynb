```python
from cs103 import *
import csv
from typing import NamedTuple, List
from enum import Enum
import matplotlib.pyplot as plt

##################
# Data Definitions

DrinkType = Enum('DrinkType', ["COFFEE", "ENERGY_DRINKS", "ENERGY_SHOTS", 
                               "SOFT_DRINKS", "TEA", "WATER"])
# interp. A type of drink, one of coffee, energy drinks, energy shots,
# soft drinks, tea, and water

# examples are redundant for enumerations

@typecheck
def fn_for_drink_type(dt: DrinkType) -> ...:
    # template from enumeration (6 fields)
    if dt == DrinkType.COFFEE:
        return ...
    elif dt == DrinkType.ENERGY_DRINKS:
        return ...
    elif dt == DrinkType.ENERGY_SHOTS:
        return ...
    elif dt == DrinkType.SOFT_DRINKS:
        return ...
    elif dt == DrinkType.TEA:
        return ...
    elif dt == DrinkType.WATER:
        return ...



DrinkData = NamedTuple('DrinkData', [('type', DrinkType),
                                    ('volume', float), # in range [10, ...)
                                    ('caffeine_amt', int)]) # in range [0, ...)

DD1 = DrinkData(DrinkType.COFFEE, 256.99, 277)
DD2 = DrinkData(DrinkType.TEA, 200.51, 50)

@typecheck
def fn_for_drink_data(dd: DrinkData) -> ...:
    # template based on compound (3 fields)
    return ...(dd.type, 
              dd.volume,
              dd.caffeine_amt)
# List[DrinkData]
# interp. a list of drink data

LODD0 = []
LODD1 = [DD1, DD2]

@typecheck
def fn_for_lodd(lodd: List[DrinkData]) -> ...:
    # template based on arbitrary-sized data and reference rule
    # description of the accumulator
    acc = ...   # type: ...

    for dd in lodd:
        acc = ...(fn_for_drink_data(dd), acc)

    return ...(acc)

```

```python
#######################
# Data definitions that MIGHT be used

# List[str]
# interp. a list of strings
LOS0 = []
LOS1 = ['hello', 'world']

# template based on arbitrary-sized data
@typecheck
def fn_for_los(los: List[str]) -> ...:
    # description of accumulator
    acc = ... # type: ...
    
    for s in los:
        acc = ...(s, acc)
        
    return ...(acc)


# List[int]
# interp. a list of integers
LOI0 = []
LOI1 = [1, -12]

# template based on arbitrary-sized data
@typecheck
def fn_for_loi(loi: List[int]) -> ...:
    # description of accumulator
    acc = ... # type: ...
    
    for i in loi:
        acc = ...(i, acc)
        
    return ...(acc)

# List[float]
# interp. a list of floats
LOI0 = []
LOI1 = [1.0, -12.0]

# template based on arbitrary-sized data
@typecheck
def fn_for_lof(lof: List[float]) -> ...:
    # description of accumulator
    acc = ... # type: ...
    
    for f in lof:
        acc = ...(f, acc)
        
    return ...(acc)
```

```python
###########
# Functions

@typecheck
def main(filename: str) -> None:
    """
    Reads the file from given filename and creates a bar graph that displays the 
    average concentration of caffeine (mg/ml) in different types of drinks.
    
    Drinks are assumed to be one or more of "Coffee", 
    "Energy Drinks", "Energy Shots", "Soft Drinks", "Tea", and "Water".
    
    Program automatically sorts the data in the order its type first appears in. 
    (e.g. when it sees energy drinks, it will calculate the total average for all energy drinks, 
    and that will be the first bar).
    """
#    return None # stub
    # Template from HtDAP, based on function composition 
    return show_bar_chart(read(filename)) 


@typecheck
def show_bar_chart(lodd: List[DrinkData]) -> None: 
    """ 
    Takes a list of drink data containing: 
    its type of drinks being one of coffee, energy drink, energy shot, soft drink, tea, and water, 
    the drink's caffeine content in mg, and the drink's volume per serving in ml.
    
    Shows a bar chart displaying the average caffeine concentration in mg/ml of each type of drink in the list.
    """ 
    #return None # stub
    
    # Template based on visualization
    
    # the width of each bar
    bar_width = 1
    
    # labels which should be ["Coffee", "Energy Drinks", "Energy Shots", "Soft Drinks", "Tea", "Water"]
    labels = get_labels(lodd)
    
    averages = get_avgs_for_labels(lodd, labels)
    
    # the opacity for the bars
    opacity = 0.8
    
    # create the chart
    rects1 = plt.bar(add_breaks(labels), # adds breaks to the labels so the long labels will have breaks at spaces
                     averages, # list containing the height for each bar
                     bar_width,
                     alpha=opacity,
                     color='1',
                     linewidth=1,
                     edgecolor="0" )                  

    # set the labels for the x-axis, y-axis, and plot title
    plt.xlabel('Drink Type')
    plt.ylabel('Avg Caffeine Concentration (mg/ml)')
    plt.title('Caffeine Concentration in Drinks')
    
    
    # set the x-coordinate for positioning the labels.
    x_coord_labels = add_breaks(labels)
    
    # set the labels for each 'tick' on the x-axis
    tick_labels = add_breaks(labels)
    
    plt.xticks(x_coord_labels, tick_labels)
    
    # show the plot
    plt.show()
  
    return None


@typecheck
def read(filename: str) -> List[DrinkData]:
    """    
    reads information from the specified file and returns a list of drink data
    """
#    return []  #stub
    # Template from HtDAP
    # lodd contains the result so far
    lodd = [] # type: List[DrinkData]

    with open(filename) as csvfile:
        
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        for row in reader:
            # you may not need to store all the rows, and you may need
            # to convert some of the strings to other types
            
            # we need the drink type(row[4]), the volume (row[1]), and caffeine (row[3])
            
            # we are going to round the decimals for the volume to 2,
            # since the scale is so small that it won't change the result
            dd = DrinkData(parse_drink_type(row[4]), round(parse_float(row[1]), 2), parse_int(row[3]))
            lodd.append(dd)
    
    return lodd


@typecheck
def parse_drink_type(s: str) -> DrinkType:
    """
    Given a string which must be one of "Coffee", "Energy Drinks", "Energy Shots",
    "Soft Drinks", "Tea", and "Water", returns the corresponding DrinkType.
    """
#    return DrinkType.COFFEE # stub

    # template based on atomic non-distinct
    if s == "Coffee":
        return DrinkType.COFFEE
    elif s == "Energy Drinks":
        return DrinkType.ENERGY_DRINKS
    elif s == "Energy Shots":
        return DrinkType.ENERGY_SHOTS
    elif s == "Soft Drinks":
        return DrinkType.SOFT_DRINKS
    elif s == "Tea":
        return DrinkType.TEA
    elif s == "Water":
        return DrinkType.WATER


@typecheck
def get_avgs_for_labels(lodd: List[DrinkData], labels: List[str]) -> List[float]:
    """
    returns the average concentrations for all drink types in the list for the specified labels
    being one of "Coffee", "Energy Drinks", "Energy Shots", "Soft Drinks", "Tea", and "Water"
    
    Function makes sure that the returned list has the averages in the order of the labels
    so if the labels were ["Tea", "Coffee"], it would return [avg for tea, avg for coffee].
    
    This way unsorted lists will still return the correct data
    """
#    return [] # stub

    # template from List[str] with extra parameter lodd
    
    # avgs_so_far contains the averages calculated so far
    avgs_so_far = []   # type: List[float]
    
    # been_calculated to track what type has already been calculated
    been_calculated = []  # type: List[str]
    
    
    for l in labels:
        if l == "Coffee" and not in_list(been_calculated, l):
            avgs_so_far.append(get_avg_concentration(lodd, DrinkType.COFFEE))
            been_calculated.append(l)
            
        elif l == "Energy Drinks" and not in_list(been_calculated, l):
            avgs_so_far.append(get_avg_concentration(lodd, DrinkType.ENERGY_DRINKS))
            been_calculated.append(l)
            
        elif l == "Energy Shots" and not in_list(been_calculated, l):
            avgs_so_far.append(get_avg_concentration(lodd, DrinkType.ENERGY_SHOTS))
            been_calculated.append(l)
            
        elif l == "Soft Drinks" and not in_list(been_calculated, l):
            avgs_so_far.append(get_avg_concentration(lodd, DrinkType.SOFT_DRINKS))
            been_calculated.append(l)
            
        elif l == "Tea" and not in_list(been_calculated, l):
            avgs_so_far.append(get_avg_concentration(lodd, DrinkType.TEA))
            been_calculated.append(l)
            
        elif l == "Water" and not in_list(been_calculated, l):
            avgs_so_far.append(get_avg_concentration(lodd, DrinkType.WATER))
            been_calculated.append(l)
            
    return avgs_so_far


@typecheck
def get_avg_concentration(lodd: List[DrinkData], drink_type: DrinkType) -> float:
    """
    Takes all the values of the drink type and returns the average concentration 
    of caffeine for that drink type (rounds to 2 decimal places).
    
    returns 0.0 if not in list
    """
#    return 0.0 # stub
    # template from List[DrinkData] with extra parameter drink_type
    
    # total contains the total concentration (all values added and divided by volume)
    total = 0.0   # type: float
    # num_added contains the amount of values that have been added so far (for averaging purposes)
    num_added = 0 # type: int
    
    for dd in lodd:
        if is_same_type(get_type(dd), drink_type):
            concentration = get_caffeine(dd) / get_volume(dd)
            total = total + concentration
            num_added = num_added + 1
    if num_added > 0:
        return round((total / num_added), 2)
    else:
        return 0.0
    

@typecheck
def get_labels(lodd: List[DrinkData]) -> List[str]:
    """
    Takes a list of drink data and returns the relevant drink type labels as a list of strings.
    
    Does not duplicate drink type labels.
    """
#    return [] # stub
    # template from List[DrinkData]
    
    # labels contains the labels seen so far (without duplicates)
    labels = []   # type: List[str]

    for dd in lodd:
        if not in_list(labels, get_drink_label(dd)):
            labels.append(get_drink_label(dd))

    return labels

@typecheck
def get_drink_label(dd: DrinkData) -> str:
    """
    Takes a DrinkData and returns the DrinkType as a string label
    """
#    return "" # stub
    # template from DrinkType
    if get_type(dd) == DrinkType.COFFEE:
        return "Coffee"
    elif get_type(dd) == DrinkType.ENERGY_DRINKS:
        return "Energy Drinks"
    elif get_type(dd) == DrinkType.ENERGY_SHOTS:
        return "Energy Shots"
    elif get_type(dd) == DrinkType.SOFT_DRINKS:
        return "Soft Drinks"
    elif get_type(dd) == DrinkType.TEA:
        return "Tea"
    elif get_type(dd) == DrinkType.WATER:
        return "Water"

@typecheck
def is_same_type(dt1: DrinkType, dt2: DrinkType) -> bool:
    """
    returns True if dt1 is the same type as dt2
    """
#    return True # stub
    # template from DrinkData with 2 parameters of DrinkData
    return dt1 == dt2

@typecheck
def get_type(dd: DrinkData) -> DrinkType:
    """
    Takes a DrinkData and returns the DrinkType
    """
#    return DrinkType.TEA # stub
    # template from DrinkData
    return dd.type

@typecheck
def get_volume(dd: DrinkData) -> float:
    """
    returns the volume in a serving of DrinkData (ml)
    """
    # return 0.0 # stub
    # template from DrinkData
    return dd.volume
    
@typecheck
def get_caffeine(dd: DrinkData) -> int:
    """
    returns the amount of caffeine in DrinkData (mg)
    """
    # return 0 # stub
    # template from DrinkData
    return dd.caffeine_amt

@typecheck
def in_list(los: List[str], d: str) -> bool:
    """
    returns True if a string dl is already in a given list; False otherwise
    """
    # template from List[str] with extra parameter dd   
    for s in los:
        if s == d:
            return True
        
    return False

@typecheck
def add_breaks(los: List[str]) -> List[str]:
    """
    Takes a los and finds spaces and creates a line break instead (Purely for aesthetic purposes).
    Returns result as a new list of strings
    """
#    return [] # stub
    # template from List[str]
    # cleaned_list contains the list that has been cleaned with line breaks
    cleaned_list = [] # type: List[str]
    
    for s in los:
        if " " in s:
            s = s.replace(" ", "\n")
        cleaned_list.append(s)
        
    return cleaned_list



#############################
# Tests for main
#############################

start_testing()

## Case 1: Empty csv with only titles

# Should display an empty plot with the title "Caffeine Concentration in Drinks", and an x axis "Drink Type",
# and y axis "Avg Caffeine Concentration (mg/ml)".
expect(main("/data/caffeine_test_empty.csv"), None) 


## Case 2: 1 row csv

# Should display a bar chartwith the title "Caffeine Concentration in Drinks", and an x axis "Drink Type",
# and y axis "Avg Caffeine Concentration (mg/ml)", and ONE bar displaying the average caffeine concentration in COFFEE
expect(main("/data/caffeine_test_1row.csv"), None)


## Case 3: csv with one of each type

# Should display a bar chartwith the title "Caffeine Concentration in Drinks", and an x axis "Drink Type",
# and y axis "Avg Caffeine Concentration (mg/ml)", and SIX bars displaying the average 
# caffeine concentration in the following drink types: Coffee, Energy Drinks, Energy Shots, Soft Drinks, Tea, Water
expect(main("/data/caffeine_test_multirow.csv"), None) 


## Case 4: unsorted csv in weird order

# Should display a bar chartwith the title "Caffeine Concentration in Drinks", and an x axis "Drink Type",
# and y axis "Avg Caffeine Concentration (mg/ml)", and SIX bars displaying the average 
# caffeine concentration in the following drink types: Coffee, Energy Drinks, Energy Shots, Soft Drinks, Tea, Water
# Will not be in this order, but should be in the order where they first appear in csv
expect(main("/data/caffeine_test_unsorted.csv"), None)

summary()


#############################
# Tests for show_bar_chart
#############################

start_testing()

# Should display an empty plot with the title "Caffeine Concentration in Drinks", and an x axis "Drink Type",
# and y axis "Avg Caffeine Concentration (mg/ml)".
expect(show_bar_chart([]), None) 

# Should display a bar chartwith the title "Caffeine Concentration in Drinks", and an x axis "Drink Type",
# and y axis "Avg Caffeine Concentration (mg/ml)", and TWO bars displaying 
# the average caffeine concentration in COFFEE, and TEA
expect(show_bar_chart([DrinkData(DrinkType.COFFEE, 256.99, 277), 
                       DrinkData(DrinkType.COFFEE, 234.8, 160), 
                       DrinkData(DrinkType.TEA, 223.01, 112), 
                       DrinkData(DrinkType.TEA, 260.22, 96)]), None)

summary()


#############################
# Tests for read
#############################

# Begin testing
start_testing()

# Examples and tests for read
expect(read("/data/caffeine_test_empty.csv"), [])
expect(read("/data/caffeine_test_1row.csv"), [DrinkData(DrinkType.COFFEE, 256.99, 277)])
expect(read("/data/caffeine_test_multirow.csv"), [DrinkData(DrinkType.COFFEE, 256.99, 277), 
                                            DrinkData(DrinkType.ENERGY_DRINKS, 250.19, 80),
                                            DrinkData(DrinkType.ENERGY_SHOTS, 57.08, 225),
                                            DrinkData(DrinkType.SOFT_DRINKS, 330.04, 43),
                                            DrinkData(DrinkType.TEA, 946.35, 100),
                                            DrinkData(DrinkType.WATER, 473.18, 60)
                                           ])
# show testing summary
summary()

#############################
# Tests for parse_drink_type
#############################


start_testing()

expect(parse_drink_type("Coffee"), DrinkType.COFFEE)
expect(parse_drink_type("Energy Drinks"), DrinkType.ENERGY_DRINKS)
expect(parse_drink_type("Energy Shots"), DrinkType.ENERGY_SHOTS)
expect(parse_drink_type("Soft Drinks"), DrinkType.SOFT_DRINKS)
expect(parse_drink_type("Tea"), DrinkType.TEA)
expect(parse_drink_type("Water"), DrinkType.WATER)

summary()

#################################
# Tests for get_avgs_for_labels
#################################

start_testing()

# test for empty lists
expect(get_avgs_for_labels([], []), [])

# test for empty data but provided labels (maybe given data but missing values in a csv file)
expect(get_avgs_for_labels([], ["Coffee", "Tea"]), [0.0, 0.0])

# test that meets conditions being given data and given labels
expect(get_avgs_for_labels([DrinkData(DrinkType.COFFEE, 256.99, 277), 
                       DrinkData(DrinkType.COFFEE, 234.8, 160), 
                       DrinkData(DrinkType.TEA, 223.01, 112), 
                       DrinkData(DrinkType.TEA, 260.22, 96)], ["Coffee", "Tea"]), [0.88, 0.44])

# test that meets conditions with more than 2 labels
expect(get_avgs_for_labels([DrinkData(DrinkType.COFFEE, 256.99, 277),
                            DrinkData(DrinkType.COFFEE, 234.8, 160),
                            DrinkData(DrinkType.ENERGY_DRINKS, 223.01, 112), 
                            DrinkData(DrinkType.TEA, 223.01, 112), 
                            DrinkData(DrinkType.TEA, 260.22, 96)], ["Coffee", "Energy Drinks", "Tea"]), [0.88, 0.5, 0.44])

# test with given data for 3 types but only 2 labels provided
expect(get_avgs_for_labels([DrinkData(DrinkType.COFFEE, 256.99, 277),
                            DrinkData(DrinkType.COFFEE, 234.8, 160),
                            DrinkData(DrinkType.ENERGY_DRINKS, 223.01, 112), 
                            DrinkData(DrinkType.TEA, 223.01, 112), 
                            DrinkData(DrinkType.TEA, 260.22, 96)], ["Coffee", "Tea"]), [0.88, 0.44])

# test that meets conditions with more than 2 labels and one being misplaced (not in order)
expect(get_avgs_for_labels([DrinkData(DrinkType.COFFEE, 256.99, 277),
                            DrinkData(DrinkType.ENERGY_DRINKS, 223.01, 112),
                            DrinkData(DrinkType.COFFEE, 234.8, 160),
                            DrinkData(DrinkType.TEA, 223.01, 112), 
                            DrinkData(DrinkType.TEA, 260.22, 96)], ["Coffee", "Energy Drinks", "Tea"]), [0.88, 0.5, 0.44])

# test that meets conditions with jumbled list
expect(get_avgs_for_labels([DrinkData(DrinkType.TEA, 260.22, 96),
                            DrinkData(DrinkType.COFFEE, 256.99, 277),
                            DrinkData(DrinkType.ENERGY_DRINKS, 223.01, 112),
                            DrinkData(DrinkType.COFFEE, 234.8, 160),
                            DrinkData(DrinkType.TEA, 223.01, 112)], ["Coffee", "Energy Drinks", "Tea"]), [0.88, 0.5, 0.44])


summary()

##################################
# Tests for get_avg_concentration
##################################
start_testing()

expect(get_avg_concentration([], DrinkType.TEA), 0.0)
expect(get_avg_concentration([DrinkData(DrinkType.COFFEE, 256.99, 277),
                              DrinkData(DrinkType.ENERGY_DRINKS, 250.19, 80)], DrinkType.TEA), 0.0)

expect(get_avg_concentration([DrinkData(DrinkType.COFFEE, 256.99, 277), 
                                            DrinkData(DrinkType.ENERGY_DRINKS, 250.19, 80),
                                            DrinkData(DrinkType.ENERGY_SHOTS, 57.08, 225),
                                            DrinkData(DrinkType.SOFT_DRINKS, 330.04, 43),
                                            DrinkData(DrinkType.TEA, 946.35, 100),
                                            DrinkData(DrinkType.WATER, 473.18, 60)], DrinkType.COFFEE), 1.08)

summary()

#############################
# Tests for get_labels
#############################

start_testing()

# test for empty list
expect(get_labels([]), [])

# test for one type
expect(get_labels([DrinkData(DrinkType.COFFEE, 256.99, 277)]), ["Coffee"])
# test for multiple types
expect(get_labels([DrinkData(DrinkType.COFFEE, 256.99, 277), 
                       DrinkData(DrinkType.TEA, 223.01, 112)]), ["Coffee", "Tea"])
# test for multipe cases of multiple types
expect(get_labels([DrinkData(DrinkType.COFFEE, 256.99, 277), 
                       DrinkData(DrinkType.COFFEE, 234.8, 160), 
                       DrinkData(DrinkType.TEA, 223.01, 112), 
                       DrinkData(DrinkType.TEA, 260.22, 96)]), ["Coffee", "Tea"])

summary()

#############################
# Tests for get_drink_label
#############################
start_testing()

expect(get_drink_label(DrinkData(DrinkType.COFFEE, 256.99, 277)), "Coffee")
expect(get_drink_label(DrinkData(DrinkType.ENERGY_DRINKS, 256.99, 277)), "Energy Drinks")
expect(get_drink_label(DrinkData(DrinkType.TEA, 256.99, 277)), "Tea")

summary()

##################################
# Tests for is_same_type
##################################

start_testing()

expect(is_same_type(DrinkType.ENERGY_DRINKS, DrinkType.ENERGY_DRINKS), True)

expect(is_same_type(DrinkType.COFFEE, DrinkType.ENERGY_DRINKS), False)

summary()

#############################
# Tests for get_type
#############################   
start_testing()

expect(get_type(DrinkData(DrinkType.COFFEE, 256.99, 277)), DrinkType.COFFEE)
expect(get_type(DrinkData(DrinkType.ENERGY_DRINKS, 256.99, 277)), DrinkType.ENERGY_DRINKS)
expect(get_type(DrinkData(DrinkType.ENERGY_SHOTS, 256.99, 277)), DrinkType.ENERGY_SHOTS)
expect(get_type(DrinkData(DrinkType.SOFT_DRINKS, 256.99, 277)), DrinkType.SOFT_DRINKS)
expect(get_type(DrinkData(DrinkType.TEA, 256.99, 277)), DrinkType.TEA)
expect(get_type(DrinkData(DrinkType.WATER, 256.99, 277)), DrinkType.WATER)

summary()

#############################
# Tests for get_volume
#############################

start_testing()

expect(get_volume(DrinkData(DrinkType.COFFEE, 256.99, 277)), 256.99)

summary()

#############################
# Tests for get_caffeine
#############################

start_testing()

expect(get_caffeine(DrinkData(DrinkType.COFFEE, 256.99, 277)), 277)

summary()


#############################
# Tests for in_list
#############################
start_testing()

# test for empty list and empty string provided
expect(in_list([], ""), False)

# test for empty list
expect(in_list([], "test"), False)

# test for list but not matching string
expect(in_list(["not test", "hello"], "test"), False)

# test where all conditions are met
expect(in_list(["not_test", "hello", "test", "try"], "test"), True)

summary()

#############################
# Tests for add_breaks
#############################

start_testing()

expect(add_breaks([]), [])
expect(add_breaks(["Coffee"]), ["Coffee"])
expect(add_breaks(["Energy Drinks"]), ["Energy\nDrinks"])
expect(add_breaks(["Coffee", "Energy Drinks", "Energy Shots", "Soft Drinks", "Tea", "Water"]), ["Coffee","Energy\nDrinks","Energy\nShots","Soft\nDrinks", "Tea", "Water"])


summary()
```
